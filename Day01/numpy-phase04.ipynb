```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 =np.random.rand(3,3)
arr3 = np.zeros((4,4))
```

```python
np.save("arrat1.npy",arr1)
np.save("arrat2.npy",arr2)
np.save("arrat3.npy",arr3)


load_arr1 = np.load("arrat1.npy")
print(load_arr1)
```

```python
try:
  logo = np.load("numpy-logo.npy")
  # Display
  plt.figure(figsize=(10,5))
  plt.title("Numpy Logo")
  plt.subplot(121)
  plt.imshow(logo)
  plt.show()

  dark_logo = 1-logo
  plt.subplot(122)
  plt.imshow(dark_logo)
  plt.show()
except FileNotFoundError:
  print("Numpy Logo not found")
```
