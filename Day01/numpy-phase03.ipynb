```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id,2021,2022,2023,2024]
sales_data = np.array([
    [1,1500000,180000,220000,250000],
    [2, 120000, 140000, 160000, 180000],
    [3, 90000, 110000, 130000, 150000],
    [4, 200000, 210000, 230000, 260000],
    [5, 80000, 95000, 110000, 125000],
    [6, 170000, 190000, 200000, 220000],
    [7, 130000, 155000, 170000, 195000]
])

print("================================")
print(sales_data.shape)
print("sample data for first 3 restaurants")
print(sales_data[:3,1:])
```

```python
# Total sales data
print(np.sum(sales_data,axis=0))

yearly_total = np.sum(sales_data[:,1:],axis=0)
print("years total \n",yearly_total)


# Minimum Sales per restaurant
min_sales = np.min(sales_data[:,1:],axis=1)
print("\n mimimum sales per restaurant \n",min_sales)


# Maximum Sales per year


max_sales = np.max(sales_data[:,1:],axis=0)
print("\n max sales per year \n",max_sales)


# average sales per restaurant

average_sales = np.mean(sales_data[:,1:],axis=1)
print("\n average sales per restaurant \n",average_sales)
```

```python
cummulative_sum = np.cumsum(sales_data[:,1:],axis=1)
print("\n cummulative sum \n",cummulative_sum)
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cummulative_sum,axis=0))
plt.title("Average Cummlative Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Average Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([5,4,3,2,1])


# Adding Vectors
print("Basic Method")
print(vector1+vector2)
print("Numpy Method")
print(np.add(vector1,vector2))


# Multiplications
print("-----------------------------")
print("Basic Method")
print(vector1*vector2)
print("Numpy Method")
print(np.multiply(vector1,vector2))


# Dot Product
print("-----------------------------")
print("Basic Method")
print(np.dot(vector1,vector2))
print("Numpy Method")
print(vector1.dot(vector2))
```
